```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

#### Задачи рекомендованные ChatGPT(плохие)

* Гистограмма возраста пассажиров: Позволит визуализировать распределение возрастов пассажиров и получить представление о возрастной структуре.

* Круговая диаграмма для CryoSleep: Покажет пропорцию пассажиров, выбравших или не выбравших сон в крио-подвешенности во время путешествия.

* Столбчатая диаграмма для HomePlanet: Позволит сравнить количество пассажиров, отправившихся с разных планет, и выявить популярные точки отправления.

* Столбчатая диаграмма для Destination: Покажет количество пассажиров, направляющихся на каждую планету, и поможет определить популярные пункты назначения.

* Линейный график или точечная диаграмма для возраста и суммарного счета в каждой из категорий аменистий (RoomService, FoodCourt, ShoppingMall, Spa, VRDeck): Позволит исследовать взаимосвязь между возрастом и использованием различных аменистий.

* Столбчатая диаграмма для VIP: Покажет количество пассажиров, выбравших или не выбравших специальное VIP-обслуживание, и поможет понять, насколько популярны такие услуги.

* Столбчатая диаграмма для Transported: Позволит оценить баланс между пассажирами, которые были транспортированы в другое измерение, и пассажирами, которые не были.

```python
df = pd.read_csv("train.csv")
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=20)
hist, bins = np.histogram(df.dropna(subset=['Age'])['Age'], bins=20)
for i in range(len(hist)):
    plt.text(bins[i], hist[i], str(int(hist[i])), va='bottom', ha="left")
plt.xticks(bins, bins.astype(int))
```

```python
plt.pie(df.dropna()['CryoSleep'].value_counts(), labels=['Нет', 'Да'], )
```

```python
plt.figure(facecolor="lightgray")
plt.pie(df['HomePlanet'].value_counts(), labels=df['HomePlanet'].value_counts().index, autopct='%1.1f%%', textprops={'color': 'white'})
```

```python
plt.bar(df['Destination'].value_counts().index, df['Destination'].value_counts().values, color=['Red', 'blue', 'green'])
```

```python
df['Destination'].unique()
```

```python
amenities_total = df.groupby('Age')[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].sum()

# Построение точечной диаграммы
plt.figure(figsize=(10, 6))
plt.scatter(amenities_total.index, amenities_total['RoomService'], label='RoomService')
plt.scatter(amenities_total.index, amenities_total['FoodCourt'], label='FoodCourt')
plt.scatter(amenities_total.index, amenities_total['ShoppingMall'], label='ShoppingMall')
plt.scatter(amenities_total.index, amenities_total['Spa'], label='Spa')
plt.scatter(amenities_total.index, amenities_total['VRDeck'], label='VRDeck')
plt.title('Использование аменистий в зависимости от возраста')
plt.xlabel('Возраст')
plt.ylabel('Суммарный счет')
plt.legend()
plt.show()
```
